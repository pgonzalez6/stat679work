```python
import time
import re
import datetime
import sys
import os.path
import csv


with open('energy.csv') as energy:
    linelist = energy.readlines()
    Energy_date= []
    Energy_value= []
    for row in linelist:
        a = re.search('^(\d+)',row)
        if a:
            date= re.sub(r'\,(\d+)',"",row)
            date= re.sub(r'\s(\d+):(\d+):(\d+)\s-(\d+)(\d+)',"",date)   
            date= re.sub(r'(\d+):(\d+):(\d+)',"",date) 
            value= re.sub(r'(\d+)-(\d+)-(\d+)\s(\d+):(\d+):(\d+)\s-(\d+)\,',"",row) 
            Energy_date.append(date.rstrip())
            Energy_value.append(value.rstrip())
        else:
            continue


new_energy_date = []
for time in Energy_date:  ### Change date format
    new_energy_date1 = datetime.datetime.strptime(time, '%Y-%m-%d').strftime('%m/%d/%y')
    new_energy_date.append(new_energy_date1)

for i in range(len(new_energy_date)-1):  ### sorted data
    if new_energy_date[i] > new_energy_date[i+1]: 
        new_energy_date.sort(key=lambda x: time.mktime(time.strptime(x,'%m/%d/%y')))
    
else:
    
    with open('waterTemperature.csv') as Temp:
                Temperature = Temp.readlines()
                List = []
                Temp_date = []
                merge = []
                n_energy = 1
                EnergyDay = new_energy_date[n_energy-1]
                for row in Temperature:
                    a = re.search(r'^(\d+)',row)
                    if a:
                        List.append(row.rstrip())

                    else:
                        continue

                for row in List:
                    date = re.sub(r'^(\d+)\,','',row)
                    date = re.sub(r'\,(\d+).(\d+)','',date)
                    date = re.sub(r'\s(\d+):(\d+):(\d+)\s(\w+)',"",date)
                    date = date.rstrip() 
                    Temp_date.append(date)    


                for i in range(len(List)-1):
                    while n_energy <= len(Energy_value):
                        rowA = List[i]
                        rowB = List[i+1] 
                        day1 = re.sub(r'^(\d+)\,','',rowB)
                        day1 = re.sub(r'\,(\d+).(\d+)','',day1)
                        day2 = re.sub(r'\s(\d+):(\d+):(\d+)\s(\w+)',"",day1)
                        day2 = day2.rstrip() 
                        while day2 >= EnergyDay:
                            if day2 == EnergyDay:
                                merge_line = rowA + "," 
                                merge.append(merge_line)
                                break
                            else:
                                energy = Energy_value[n_energy]
                                energy_1000 = float(energy)/1000 
                                merge_line = rowA +","+ str(energy_1000)
                                merge.append(merge_line)
                                n_energy += 1 
                                EnergyDay = new_energy_date[n_energy - 1] 
                                break
                        break
                    merge.append(List[len(List) -1]+',')
                    header_list = ['Row_line',',','Date Time',',','Temperature',',','Energy(Wh/1000)']
                file_out = sys.stdout
                if file_out == sys.stdout: 
                    for i in range(len(header_list)):
                        file_out.write(header_list[i])
                        file_out.write('\n')
                    for j in merge: 
                        file_out.write(j)
                        file_out.write('\n')
                                              
                                    
                with open(file_out,'Temp') as output:
                                        for i in range(len(header_list)):
                                            output.write(header_list[i])
                                            output.write('\n')
                                       
                                        for k in range(len(merge)):
                                            output.write(merge[k])
                                            output.write('\n')    

print(output)
```
